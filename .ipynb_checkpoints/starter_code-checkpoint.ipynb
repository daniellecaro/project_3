```python
# Group the food access df by county to prep for df merge
grouped_df = food_access_atlas.groupby(food_access_atlas["County"]).sum()
grouped_df
```

```python
# Merge dfs on "County"
merged_df = pd.merge(diabetes_atlas, grouped_df, on=["County"])
merged_df
```

```python
# set condition to filter only the necessary states
options = ['Texas', 'Virginia', 'California'] 
    
# selecting rows based on condition 
filtered_df = merged_df[merged_df['State'].isin(options)] 
filtered_df
```
